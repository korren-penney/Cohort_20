```python
# The purpose of this program is to use the random module in python 
# to roll a dice for our icebreaker today. Korren is going to attempt
# to do this live, be prepared for typos.

# import our random library
import random

# declare our variables to determine the smallest and larges numbers
min = 1
max = 6

# declare a var to keep track of if we want to roll again
# set the var to yes
roll_again = "yes"

# start our loop
while roll_again == 'yes':
    # print a message to indicate the die is rolling
    print("Rolling the die....")
    
    # declare var for our die value and assign random number
    die_value = (random.randint(min,max))
    
    # print the die value
    print("The die value is: ", die_value)
    
    # create the if statement to tell the program what to do for each 
    # potential number
    if die_value == 1:
        print("If you were invisible for a day, what would you do?")
    elif die_value == 2:
        print("If you could choose a super power, what would it be?")
    elif die_value == 3:
        print("What is your favourite food?")
    elif die_value == 4:
        print("Who is your favourite book or movie character?")
    elif die_value == 5:
        print("What is your favourite game to play?")
    else:
        print("what is one thing you are good at?")
        
    # prompt user to se if they want to roll again
    roll_again = input("Roll again? yes/no ")
    
print("This code runs only at the end! Good bye")
```
